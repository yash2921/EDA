```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
taxis=sns.load_dataset('taxis')
```

```python
taxis.head()
```

```python
sns.scatterplot(data=taxis, x='tip', y='distance',color='green',marker='*')
plt.title('Distance per Tips')
plt.show()
```

**The plot shows that more the distance will have more tip**

```python
sns.scatterplot(data=taxis, x='distance', y='fare')
plt.show()
```

**The plot shows that more the distance will have more fare**

```python
payment_mode=taxis.groupby('payment')['total'].sum()
payment_mode
```

```python
payment_type=taxis['payment'].unique()[0:2]
payment_type
```

```python
plt.pie(payment_mode,labels=payment_type,autopct='%1.1f%%',explode=[0,0.1],shadow=True)
plt.show()
```

**The chart show that by which mode through payment is done more**



```python
taxis.groupby('pickup_borough')['total'].sum().sort_values(ascending=False)
```

```python
taxis.groupby('pickup_borough')['passengers'].sum().sort_values(ascending=True).plot(kind='bar')
```

**Maximum Revenue Generate from which pickup_borough and also passengers bookings**

```python
taxis['dropoff_month']=taxis['dropoff'].dt.month
taxis['dropoff_date']=taxis['dropoff'].dt.date
taxis['dropoff_hour']=taxis['dropoff'].dt.hour
taxis['pickup_day']=taxis['pickup'].dt.day_name()
taxis['dropoff_day']=taxis['dropoff'].dt.day_name()
```

```python
taxis.head()
```

```python
sns.countplot(x="pickup_day",data=taxis)
```

**We can clearly see that Saturday and Friday seem to be the busiest days for Pickup**

```python
sns.countplot(x="dropoff_day",data=taxis)
```

**We can clearly see that Saturday and Friday seem to be the busiest days for Dropoff**

```python
sns.catplot(x="passengers", y='total' ,data=taxis)
sns.catplot(x="passengers", y='tip', data=taxis)
```

**We can see the tip amount is high when passenger count in the trip is just one.**

**We can also see that the total amount paid is high when passenger count is one.**

```python
sns.lineplot(x="pickup_day",y='tip',data=taxis)
```

```python
sns.lineplot(x="pickup_day",y='total',data=taxis)
```

**We can see that the tip amount is highest on wednesday and lowest on sunday for pickup days and also the total amount is also highest on wednesday while it is lowest on sunday. So it is better to run less trips on sunday to increase the profits.**

```python
ax1=taxis.groupby("passengers")[['fare', 'tip','tolls', 'total']].mean().reset_index()
```

```python
ax1.plot(kind='bar')
```

**We see that overall taxi costs are fairly uniform across passenger count. Since taxi fares are based on meters that calculate trip distance and time, this makes sense!**
